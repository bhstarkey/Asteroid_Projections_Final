```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import requests
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the website
url = 'http://www.passc.net/EarthImpactDatabase/New%20website_05-2018/World.html'

browser.visit(url)
```

```python
# Create a Beautiful Soup Object
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
links = html_soup.find_all('a')

for link in links:
    print(link) 
```

```python
pages = ['NorthAmerica.html',
       'SouthAmerica.html',
       'Europe.html',
       'AsiaRussia.html',
       'Africa.html',
       'Australia.html']
base_url = 'http://www.passc.net/EarthImpactDatabase/New%20website_05-2018/'
```

```python
impact_crater_list = []
crater_data = pd.DataFrame(index=None)

# Loop through the scraped data to create a list of rows
for page in pages:
    URL2 = base_url + str(page)
    browser.visit(URL2)
    html = browser.html
    html_soup = soup(html, 'html.parser')
    table = html_soup.find('table', { "border": 1})
    rows = table.find_all('tr')
    for row in rows:
    #    print(row)
        col_list = []
        for col in row.find_all('td'): 
            col_list.append(col.text)
        crater_data = crater_data.append(pd.Series(col_list),ignore_index=True)
    crater_data
```

```python
browser.quit()
```
